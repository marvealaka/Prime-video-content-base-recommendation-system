```python
import numpy as np # linear algebra
import pandas as pd # data processing
```

```python
movie = pd.read_csv('C:/Users/BRIGHT WORLD/Downloads/amazon_prime_titles.csv')
movie.head()
```

```python
movie.columns
```

```python
movie[['show_id', 'type', 'title', 'director', 'listed_in', 'description']]   
```

```python
movie['tags'] = movie['type']+movie['director']+movie['listed_in']+movie['description']
```

```python
new_data = movie.drop(columns=['type', 'director', 'listed_in', 'description', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration',])
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=9668, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]: 
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movie):
    index=new_data[new_data['title']==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]: 
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Outpost")
```
